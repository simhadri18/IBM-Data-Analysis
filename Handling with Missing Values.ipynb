# replace

```python
import pandas as pd
df=pd.DataFrame([[-9999,-8888,'rainy'],
              [43,26],
              [-9999,-9999,'sunny']],columns=['Temperature','Humidity','Event'])
df
```

```python
import numpy as np
df.fillna(np.NaN,inplace=True)
df
```

```python
df.replace(-9999,0)
```

```python
df.replace({'Temperature':-9999,
            'Humidity':[-9999,-8888],
            'Event':0},
          np.NaN)
```

```python
df.replace({'Event':np.NaN},'sunny')
```

```python
df.replace({-9999:0,
            -8888:np.NaN})
```

```python
import pandas as pd
df=pd.DataFrame([['-9999f','-8888kmp','rainy'],
              [43,26],
              [-9999,-9999,'sunny']],columns=['Temperature','Humidity','Event'])
df
```

```python
new=df.replace('[A-Za-z]','',regex=True)
new
```

```python
new=df.replace({'Temperature':'[A-Za-z]',
                'Humidity':'[A-Za-z]'},'',regex=True)
new
```
